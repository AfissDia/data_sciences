```python
# Importation des bibliothèques
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as chromeService
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(service=chromeService(ChromeDriverManager().install()))
```

```python
url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
driver.get(url)
```

```python
#data = requests.get(standings_url)
```

```python
from bs4 import BeautifulSoup
```

### <span style="color:darkblue"># pour recuperer la 1ere page html concernat la saison 2019_2020</span> .


```python
soup = BeautifulSoup(driver.page_source,"html.parser")
```

```python
# recupéerer le contenu de la table pour tous les equipes de ta primer league 
standings_table = soup.select('table.stats_table')[0]
standings_table
```

### <span style="color:darkblue"># Recuperer les balises "href" des equipes du championnat</span> .

```python
links = standings_table.find_all('a')
```

```python
links = [l.get("href") for l in links]
links = [l for l in links if '/squads/' in l]
```

```python
links
```

### <span style="color:darkblue"># Recuperer les URL des équipes </span> .

```python
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

### <span style="color:darkblue"># Focus sur  team_urls[3] de mon choix: l'équipe de CHELSEA </span> .

```python
team_urls[3]
```

```python
team_name = team_urls[3].split("/")[-1].replace("-Stats", "").replace("-", " ")
team_name
```

```python
import requests
```

```python
data = requests.get(team_urls[3])
```

```python
data
```

### <span style="color:darkblue"># Recupérer le dataFrame de : l'équipe de CHELSEA saison 2019_2020</span> .

```python
matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
matches['Team']=team_name
matches
```

```python
matches_Chelsea = matches[matches['Comp']=='Premier League']
matches_Chelsea
```

```python
matches_Chelsea.shape
```

### <span style="color:darkblue"># Recupérer le dataFrame de : l'équipe de CHELSEA saison 2019_2020 dans un fichier csv</span>.

```python
matches_Chelsea.to_csv("df_matches_Chelsea_PL_2019_2020.csv")
```
